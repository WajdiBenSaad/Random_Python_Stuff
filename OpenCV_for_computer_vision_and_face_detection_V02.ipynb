```python
#Loading the image to be tested
test_image = cv2.imread('img/bb.jpg')

#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
```

```python
# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
#Since we know that OpenCV loads an image in BGR format, so we need to convert it into RBG format to be able to display its true colors. Let us write a small function for that.
```

```python
#OpenCV loads an image in BGR format, so we need to convert it into RBG 
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
#Loading the classifier for frontal face 
haar_cascade_face = cv2.CascadeClassifier('C:\\Users\obensaad.ext\\OneDrive - LOUVRE HOTELS Group\\Louvre Hotels\\Python Scripts\\data\\haarcascades\\haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
##Our next step is to loop over all the coordinates it returned and draw rectangles around them using Open CV.
##We will be drawing a green rectangle with a thickness of 2

for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
##Finally, we shall display the original image in colored to see if the face has been detected correctly or not.

#convert image to RGB and show image

plt.imshow(convertToRGB(test_image))
```
